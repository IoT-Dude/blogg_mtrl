### Matplotlib and visualization
Matplotlib is a visualization library in Python for 2D plots of arrays. A visualization library built on NumPy arrays.

Visualization in all kinds of data analysis is a good way to see "the lay of the land" - it allows us to visually see interesting 
datapoints in droves of data, and from that one can make further descisions on how to navigate in the material at hand. 

Matplotlib consists of several plots like line, bar, scatter, histogram etc. Like this:

### Line plot

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([2, 9, 1, 15])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

### Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Smörgåsar", "Ostskivor", "Sillfileér", "Pilsner"]) # Sandwich, cheese slices, pickled herring and pilsner.
y = np.array([3, 9, 11, 4])

plt.bar(x,y)
plt.show()
```

### Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

### Histogram

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

Example: Say you ask for the height of 250 people, you might end up with a histogram like this:

```python
import numpy as np

# Create 250 random normal distribution numbers (for example height in cm, where median is 170 and each chunk is 10 cm).
x = np.random.normal(170, 10, 250)

print(x) 
```

#### The Magic of two rows of code
The result will be printed out as random result like this, and as a visualization it look like this:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

QED, it is much less difficult to visualize data compared to only se numbers in rows and columns.

### Seaborn

Seaborn is, compared to matplotlib, another visualization library for statistical graphics plotting in Python. 
It provides useful "styles" and color palettes to make statistical plots more attractive for the viewer. 
<br>
Seaborn is built on the top of matplotlib library, and is also closely integrated to pandas.

A difference is that Seaborn is more fokused on explorative data analysis of data(as a superset of matplotlib).



Regression plots: The regression plots in seaborn are primarily intended to add a visual guide that helps to emphasize patterns in a dataset during exploratory data analyses.
Matrix plots: A matrix plot is an array of scatterplots.
Multi-plot grids: It is an useful approach is to draw multiple instances of the same plot on different subsets of the dataset.

### Plots using seaborn

#### Dist plot
Seaborn dist plot is used to plot a histogram, and for examining univariate and bivariate distributions etc

```python
import numpy as np
import seaborn as sns

# Create 250 random normal distribution numbers (for example height in cm, where median is 170 and each chunk is 10 cm).
x = np.random.normal(170, 10, 250)

sns.histplot(x)
```

#### Line plot
The line plot is one of the most basic plot in seaborn library. This plot is mainly used to visualize the data in form of some time series, i.e. in continuous manner.

```python
import pandas as pd
import seaborn as sns

Year = [2012, 2014, 2016, 2020, 2022, 2018, 2023]
Profit = [80, 55.8, 74, 65, 99.5, 19, 43.6]

data_plot = pd.DataFrame({"Year":Year, "Profit":Profit})

sns.lineplot(x = "Year", y = "Profit", data=data_plot)
plt.show()
```

#### <b>lmplot</b>
The lmplot (Linear Model Plot) is another most basic plot. It shows a line representing a linear regression model along with data points on the 2D-space, and x and y can be set as the horizontal and vertical labels respectively

It is a multi-plot grid for plotting conditional relationships. This class maps a dataset onto multiple axes arrayed in a grid of rows and columns that correspond to levels of variables in the dataset. The plots it produces are often called “lattice”, “trellis”, or “small-multiple” graphics.
<br>
It can also represent levels of a third variable with the hue parameter, which plots different subsets of data in different colors(the hue parameter determines which column in the data frame should be used for colour encoding). This uses color to resolve elements on a third dimension, but only draws subsets on top of each other and will not tailor the hue parameter for the specific visualization the way that axes-level functions that accept hue will.

So, lmplots are basically a scatterplott with overlaid regression lines - a combination of linear regression fit and an aid in understanding the relationshiop between multiple variables separately within subsets of data.


```python
import seaborn as sns
sns.set(style="ticks")

df = sns.load_dataset("anscombe")

sns.lmplot(x = 'x', y = 'y', data=df, hue='dataset')
```
